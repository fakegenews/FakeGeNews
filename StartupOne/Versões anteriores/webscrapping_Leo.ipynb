```
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import json
import urllib.parse
```

```
#String da busca
busca = 'Lula traição desonesto Pará'

#Dividindo em array acredito que quando pegar as palavras do texto virá em array
arrayBusca = busca.split(' ')

#url padrão para busca do G1
urlG1 = "https://g1.globo.com/busca/?q="

for word in arrayBusca:
  #fazendo parse das palavras de busca no formato aceito em url
  urlG1 = urlG1 + urllib.parse.quote(word) + "+"
  
#removendo o "+" que fica sobrando no fim
urlG1 = urlG1[:urlG1.rfind('+')]
urlG1
```

```
#SELENIUM
option = Options()
option.headless = True

#desta forma acontece em background
#driver = webdriver.Firefox(options=option)
driver = webdriver.Firefox(executable_path = "/Shared with me/StartupOne/geckodriver/geckodriver.exe")

driver.get(urlG1)

time.sleep(10)

driver.quit()
```
